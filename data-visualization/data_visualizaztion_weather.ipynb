```python
dados = pd.read_csv('monitoramento_tempo.csv')
dados.head()
```

```python
#analisando informações sobre o dataframe

dados.info()
```

```python
#convertendo data de objeto para datetime

dados['data'] = pd.to_datetime(dados['data'])
dados['data'].head()
```

```python
plt.plot(dados['data'] , dados['temperatura'])
```

```python
plt.figure(figsize=(20,8))
plt.plot(dados['data'], dados['temperatura'])

# 
plt.title("Variação de temperatura em função dos anos")
```

```python
#criando nova figura
```

```python
grafico = plt.figure(figsize = (15,8))
eixo = grafico.add_axes([0,0,1,1])  #  [xmin, ymin, dx, dy] 

eixo.plot(dados['data'] , dados['temperatura'] , color = 'crimson')

# dando nome aos eixos = 
eixo.set_title('Temperatura ' , fontsize = 25) # titulo
eixo.set_ylabel('Temperatura (K)',fontsize = 16)
eixo.set_xlabel('Ano' , fontsize = 16)

# adicinando legenda 
eixo.legend(['temperatura'],loc = 'upper right' , fontsize = 16)
```

```python
import datetime
```

```python
grafico = plt.figure(figsize = (15,8))
eixo = grafico.add_axes([0,0,1,1])  #  [xmin, ymin, dx, dy] 

eixo.plot(dados['data'] , dados['temperatura'] , color = 'crimson' , lw = .5 , marker = 'o')

# dando nome aos eixos = 
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))

eixo.set_title('Temperatura ' , fontsize = 25) # titulo
eixo.set_ylabel('Temperatura (K)',fontsize = 16)
eixo.set_xlabel('Ano' , fontsize = 16)

# adicinando legenda 
eixo.legend(['temperatura'],loc = 'upper right' , fontsize = 16)
```

```python
grafico = plt.figure(figsize = (15,8))
eixo = grafico.add_axes([0,0,1,1])  #  [xmin, ymin, dx, dy] 

eixo.plot(dados['data'] , dados['temperatura'] , color = 'crimson' , lw = .5)

# limitando o foco em X do grafico
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))

# dando nome aos eixos = 
eixo.set_title('Temperatura ' , fontsize = 25) # titulo
eixo.set_ylabel('Temperatura (K)',fontsize = 16)
eixo.set_xlabel('Ano' , fontsize = 16)

eixo.grid(True)

# adicinando legenda 
eixo.legend(['temperatura'],loc = 'upper right' , fontsize = 16)
```

```python
fig = plt.figure(figsize = (15,8))

eixo = fig.add_axes([0,0,1,1])
eixo_2 = fig.add_axes([0.6 , 0.6 , 0.3 , 0.3])

eixo.plot(dados['data'] , dados['temperatura'] , color = 'crimson')
eixo_2.plot(dados['data'],dados['temperatura'] , lw = 0.4 )
eixo_2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))

eixo.set_title('Temperatura ' , fontsize = 25) # titulo
eixo.set_ylabel('Temperatura (K)',fontsize = 16)
eixo.set_xlabel('Ano' , fontsize = 16)

eixo_2.set_title('Temperatura 2014-2015 ' , fontsize = 12) # titulo

eixo.legend(['temperatura'],loc='best' , fontsize = 12)

eixo.grid(True)
eixo_2.grid(True)

# [distancia do x , distancia do y , largura , altura]
```

```python
fig = plt.figure(figsize = (15,8))

eixo = fig.add_axes([0,0,1,1])
eixo_2 = fig.add_axes([0.6 , 0.7 , 0.3 , 0.2])

eixo.plot(dados['data'] , dados['temperatura'] , color = 'crimson')
eixo_2.plot(dados['data'],dados['temperatura'] , lw = 0.4 )

eixo.set_xlim(datetime.datetime(2014,6,1),datetime.datetime(2014,7,1))
eixo_2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_ylim(270,320)

eixo.set_title('Temperatura em Junho de 2014' , fontsize = 25) # titulo
eixo.set_ylabel('Temperatura (K)',fontsize = 16)
eixo.set_xlabel('Ano' , fontsize = 16)

eixo_2.set_title('Temperatura 2014-2015 ' , fontsize = 12) # titulo

eixo.legend(['temperatura'],loc='best' , fontsize = 12)

eixo.grid(True)
eixo_2.grid(True)

# [distancia do x , distancia do y , largura , altura]
```

```python
fig = plt.figure(figsize = (15,8))

eixo = fig.add_axes([0,0,1,1])
eixo_2 = fig.add_axes([0.6 , 0.7 , 0.2 , 0.2])

eixo.plot(dados['data'] , dados['temperatura'] , color = 'crimson')
eixo_2.plot(dados['data'],dados['temperatura'] , lw = 0.4 , color = 'crimson')


azul_esquerda = dados['data'] < datetime.datetime(2014,6,1)
azul_direita = dados['data'] > datetime.datetime(2014,7,1)
# define os limites horizontais para a o eixo data. 

eixo_2.plot(dados[azul_esquerda]['data'] , dados[azul_esquerda]['temperatura'], color = 'blue')
eixo_2.plot(dados[azul_direita]['data'] , dados[azul_direita]['temperatura'], color = 'blue')

eixo.set_xlim(datetime.datetime(2014,6,1),datetime.datetime(2014,7,1))
eixo_2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_ylim(270,320)

eixo.set_title('Temperatura em Junho de 2014' , fontsize = 25) # titulo
eixo.set_ylabel('Temperatura (K)',fontsize = 16)
eixo.set_xlabel('Ano' , fontsize = 16)

eixo_2.set_title('Temperatura 2014-2015 ' , fontsize = 12) # titulo

eixo.legend(['temperatura'],loc='best' , fontsize = 12)

eixo.grid(True)
eixo_2.grid(True)

# [distancia do x , distancia do y , largura , altura]
```

```python
# fig + tamanho dos eixos
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])

# plotando grafico ============================================
eixo.plot(dados['data'],dados['temperatura'], color = 'darkcyan')

# configuração estética dos eixos  =========================
eixo.set_title('Temperatura no momento' , fontsize = 25 , pad = 20)
eixo.set_xlabel('Data' , fontsize = 20)
eixo.set_ylabel('Temperatura' , fontsize = 20)

# ==========================================================
eixo.legend(['Temperatura'] , loc = 'best' , fontsize = 15)
eixo.grid(True)

# linhas limitadoras horizontais
eixo.axhline(max(dados['temperatura']), color = 'k', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--')

# anotações ================================

x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados['temperatura'])

x2 = dados['data'][dados['temperatura'].idxmax()-7000]
y2 = max(dados['temperatura']) - 5

eixo.annotate("Máximo",xy=(x1,y1),fontsize = 20, xytext = (x2,y2),
              arrowprops = dict(facecolor='k'))


x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmin() - 7000]
y2 = min(df['temperatura']) + 5

eixo.annotate('Mínimo', xy=(x1,y1), fontsize=20, 
              xytext=(x2,y2), arrowprops=dict(facecolor='k'))
```
